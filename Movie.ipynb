```python
import gradio as gr
import requests

def get_movie_suggestions(movie_name):
    url = "https://online-movie-database.p.rapidapi.com/auto-complete"
    querystring = {"q": movie_name}
    headers = {
        "X-RapidAPI-Key": "0a7b57a03emsh900e200e7213c95p1fff42jsn99287d9b2e90",
        "X-RapidAPI-Host": "online-movie-database.p.rapidapi.com"
    }
    response = requests.request("GET", url, headers=headers, params=querystring)
    if response.status_code == 200:
        return response.json()
    else:
        return "Error fetching data"

iface = gr.Interface(
    fn=get_movie_suggestions,
    inputs=gr.inputs.Textbox(label="ENTER MOVIE OR SERIES NAME HERE"),
    outputs=gr.outputs.Textbox()
)

iface.launch()
```
